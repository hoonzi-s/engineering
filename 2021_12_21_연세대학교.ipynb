```python
from bs4 import BeautifulSoup # 파이썬으로 HTML을 다루는 라이브러리
import requests # 요청하다
import pandas as pd
import time
```

```python
total = []

for n in range(1, 101):

  url = requests.get("https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query={}%ED%9A%8C%EB%A1%9C%EB%98%90".format(n)) # 네이버 로또 검색
  html = BeautifulSoup(url.text)

  numbers = html.find('div', class_ = 'num_box').find_all('span', class_ = 'num')

  box = []

  for i in numbers:
    box.append(int(i.text))

  total.append(box)

  print('{}회 로또 데이터 저장완료... {}'.format(n, box))
  
  time.sleep(0.3)
```
