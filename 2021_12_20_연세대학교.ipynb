```python
# 반복문

# while : 조건만 맞으면 무한반복 // 몇번 반복해야할지 모를때
# for : 순서가 있는 유한반복 // 몇번 돌려야 알 때

# while 조건:

hit = 0
while hit < 10:
  hit = hit + 1
  print('나무를 {}번 찍었습니다.'.format(hit))
  if hit == 10:
    print('나무가 쓰러졌습니다.')
```

```python
coffee = 10

while True:
  coffee = coffee - 1
  print('커피를 한잔 팔고, {}잔 남았습니다.'.format(coffee))
  if coffee == 0:
    print('커피를 모두 팔았습니다.')
    break # 반복문을 돌리다가 얘를 만나면, 반복문을 강제로 나갑니다.
```

```python
# 5! = 120

def fac(n):
  a = 1
  for i in range(1, n+1):
    a = a * i
  return a
```

```python
fac(5)
```

```python
# 재귀함수 : 내가 나를 부르는 함수

def fac(n):
  
  if n == 1:
    return 1
  else:
    return n * fac(n-1)
```

```python
# 퀴즈 : 피보나치 수열

# 1 1 2 3 5 8 13 21...

count = 0

dic = {1:1, 2:1}

def fibo(n):
  
  global count
  count = count + 1

  if n in dic:
    return dic[n]
  else:
    dic[n] = fibo(n-1) + fibo(n-2)
    return dic[n]
```

```python
fibo(35)
```

```python
# 클래스 : 붕어빵을 만드는 틀
# 클래스 : 함수들을 모아놓은 주머니

class Bank:
  
  def __init__(self): # 권한을 줄 때, 자동으로 한번 실행되는 함수
    print('가입을 축하합니다.')
    self.money = 0
    self.location = '초보자마을'

  def deposit(self, a): # 클래스 안에서 함수를 만들때 self 재료를 넣어야함
    self.money = self.money + a
```

```python
person1 = Bank() # person1은 Bank클래스를 자유롭게 사용할 수 있는 권한을 받은 것
```

```python
person1.deposit(50000)
```

```python
person1.money
```

```python
person2 = Bank()
person2.deposit(30000)
```

```python
person2.money
```

```python
person1.money
```

```python
person3 = Bank()
```

```python
person3.money
```
